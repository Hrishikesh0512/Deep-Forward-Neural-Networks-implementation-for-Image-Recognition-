## Task 3 Construct an input pipeline

*(weight ~15%)*

Creat train/validate/test data splits and construct tf.data pipelines. Make sure that the training data is batched. 

- How do you determine the batch size?
- Do we need to shuffle the training data? If yes, how do you determine the buffer size?


**Answers**

1)How do you determine the batch size?

Larger batch sizes result in quicker training development, but they do not always converge as quickly. Smaller batch sizes train more slowly, but they can converge more quickly. It is unquestionably problem-specific.
To a point, the models develop as the number of training epochs increases. As they converge, their accuracy will begin to plateau.

2)Do we need to shuffle the data?,If yes how do you determine the buffer size?

Yes,We do need to shuffle the data
If the buffer size is set to 100, Tensorflow will hold a buffer of the next 100 samples and pick one at random from it. The next element is then added to the buffer.
As a result, if buffer size = 1, no shuffle occurs, while if buffer size > data set size, a perfect uniform random shuffle occurs.
It is recomendeded to shuffle the data before generating the TFRecords and use a smaller buffer size are suggested.


```python
#Construting pipeline
#Spliting the data
from sklearn.model_selection import train_test_split
```

```python
#Splitting train & test dataset
X_TST, x_valid, Y_TST, y_valid = train_test_split(X_test, Y_test, test_size = 0.2, random_state = 0)
```

```python
#Reshaping dataset
temp_1 = X_train.reshape(60000,28*28)
temp_1 = (temp_1.astype('float32') / 255)
temp_imn = X_TST.reshape(X_TST.shape[0],784)
temp_imn = (temp_imn.astype('float32') / 255)
val_imn = x_valid.reshape(x_valid.shape[0],784)
val_imn= (val_imn.astype('float32') / 255)
```

```python
#Converting to catagorical data
trained_Cat = to_categorical(Y_train)
tested_Cat = to_categorical(Y_TST)
valid = to_categorical(y_valid)
l1 = len(X_train)
#Taking Inputs for testing & validating data
testing_data = tf.data.Dataset.from_tensor_slices((temp_1, trained_Cat)).shuffle(l1).batch(128) 
validating_data = tf.data.Dataset.from_tensor_slices((val_imn, valid )).batch(128) 
testing_data = tf.data.Dataset.from_tensor_slices((temp_imn, tested_Cat )).batch(128)
```
